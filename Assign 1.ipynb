```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import cv2
import os 
from tqdm import tqdm 
import os
print(os.listdir("train"))
```

```python
train_ravel = "train/Raveling"
train_nravel = "train/non_raveling"
test_data1 = "test"
```

```python
def train_data():
    train_data_ravel = [] 
    train_data_nravel= []
    for image1 in tqdm(os.listdir(train_ravel)): 
        path = os.path.join(train_ravel, image1)
        img1 = cv2.imread(path) 
        train_data_ravel.append(img1) 
    for image2 in tqdm(os.listdir(train_nravel)): 
        path = os.path.join(train_nravel, image2)
        img2 = cv2.imread(path) 
        train_data_nravel.append(img2) 
    
    train_data= np.concatenate((np.asarray(train_data_ravel),np.asarray(train_data_nravel)),axis=0)
    return train_data 
```

```python
train = train_data()
train= (train-np.min(train))/(np.max(train)-np.min(train))
# train= (train-np.mean(train))/(np.std(train))
train.shape
```

```python
def test_data():
    test_data = [] 
    for image1 in tqdm(os.listdir(test_data1)): 
        path = os.path.join(test_data1, image1)
        img1 = cv2.imread(path) 
        test_data.append(img1) 
    test_data=np.asarray(test_data)
    return test_data
```

```python
test=test_data()
test= (test-np.min(test))/(np.max(test)-np.min(test))
test.shape
```

```python
# Labelling the data

z = np.zeros(351)
o = np.ones(349)
Y_train = np.concatenate((o, z), axis=0)
Y_train=Y_train.reshape(train.shape[0],1)   ###
Y_train.shape
```

```python
# splitting training data into training and validation data

from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train, Y_train, test_size=0.10, random_state=42) ###
number_of_train = x_train.shape[0]
number_of_val = x_val.shape[0]
print(y_train.shape)
print(x_val.shape)
x_train.shape
```

```python
x_train_flatten = x_train.reshape(number_of_train,x_train.shape[1]*x_train.shape[2]*x_train.shape[3])
x_val_flatten = x_val.reshape(number_of_val,x_val.shape[1]*x_val.shape[2]*x_val.shape[3])
print("X train flatten",x_train_flatten.shape)
print("X val flatten",x_val_flatten.shape)
```

```python
x_test_flatten = test.reshape(test.shape[0],test.shape[1]*test.shape[2]*test.shape[3])
x_test = x_test_flatten.T
x_test.shape
```

```python
print(x_train.shape)
x_train[0].shape
```

```python
# LOGISTIC REGRESSION #
#  getting features
#1. mean
xm11 = ()
xm12 = ()
xm13 = ()
for i in range(x_train.shape[0]):
    xm11 = np.append(xm11,np.mean(x_train[i,:,:,0]))
    xm12 = np.append(xm12, np.mean(x_train[i,:,:,1]))
    xm13 = np.append(xm13, np.mean(x_train[i,:,:,2]))
xm11=xm11.reshape(1,xm11.shape[0])
xm12=xm12.reshape(1,xm12.shape[0])
xm13=xm13.reshape(1,xm13.shape[0])
print(xm12.shape)

xmv1 = ()
xmv2 = ()
xmv3 = ()
for i in range(x_val.shape[0]):
    xmv1 = np.append(xmv1,np.mean(x_val[i,:,:,0]))
    xmv2 = np.append(xmv2, np.mean(x_val[i,:,:,1]))
    xmv3 = np.append(xmv3, np.mean(x_val[i,:,:,2]))
xmv1=xmv1.reshape(1,xmv1.shape[0])
xmv2=xmv2.reshape(1,xmv2.shape[0])
xmv3=xmv3.reshape(1,xmv3.shape[0])
print(xmv2.shape)

xmt1 = ()
xmt2 = ()
xmt3 = ()
for i in range(test.shape[0]):
    xmt1 = np.append(xmt1, np.mean(test[i,:,:,0]))
    xmt2 = np.append(xmt2, np.mean(test[i,:,:,1]))
    xmt3 = np.append(xmt3, np.mean(test[i,:,:,2]))
xmt1=xmt1.reshape(1,xmt1.shape[0])
xmt2=xmt2.reshape(1,xmt2.shape[0])
xmt3=xmt3.reshape(1,xmt3.shape[0])
print(xmt2.shape)
```

```python
#2. standard deviation
xs1 = ()
xs2 = ()
xs3 = ()
for i in range(x_train.shape[0]):
    xs1 = np.append(xs1, np.std(x_train[i,:,:,0]))
    xs2 = np.append(xs2, np.std(x_train[i,:,:,1]))
    xs3 = np.append(xs3, np.std(x_train[i,:,:,2]))
xs1=xs1.reshape(1,xs1.shape[0])
xs2=xs2.reshape(1,xs2.shape[0])
xs3=xs3.reshape(1,xs3.shape[0])
print(xs2.shape)

xsv1 = ()
xsv2 = ()
xsv3 = ()
for i in range(x_val.shape[0]):
    xsv1 = np.append(xsv1, np.std(x_val[i,:,:,0]))
    xsv2 = np.append(xsv2, np.std(x_val[i,:,:,1]))
    xsv3 = np.append(xsv3, np.std(x_val[i,:,:,2]))
xsv1=xsv1.reshape(1,xsv1.shape[0])
xsv2=xsv2.reshape(1,xsv2.shape[0])
xsv3=xsv3.reshape(1,xsv3.shape[0])
print(xsv2.shape)

xst1 = ()
xst2 = ()
xst3 = ()
for i in range(test.shape[0]):
    xst1 = np.append(xst1, np.std(test[i,:,:,0]))
    xst2 = np.append(xst2, np.std(test[i,:,:,1]))
    xst3 = np.append(xst3, np.std(test[i,:,:,2]))
xst1=xst1.reshape(1,xst1.shape[0])
xst2=xst2.reshape(1,xst2.shape[0])
xst3=xst3.reshape(1,xst3.shape[0])
print(xst2.shape)
x_train[0,:,:,0].shape
```

```python
#3. skewness 
import scipy

xsk1 = ()
xsk2 = ()
xsk3 = ()
for i in range(x_train.shape[0]):
    xsk1 = np.append(xsk1,scipy.stats.skew(x_train[i,:,:,0].flatten(),axis=0,bias=True))
    xsk2 = np.append(xsk2,scipy.stats.skew(x_train[i,:,:,1].flatten(),axis=0,bias=True))
    xsk3 = np.append(xsk3,scipy.stats.skew(x_train[i,:,:,2].flatten(),axis=0,bias=True))
xsk1=xsk1.reshape(1,xsk1.shape[0])
xsk2=xsk2.reshape(1,xsk2.shape[0])
xsk3=xsk3.reshape(1,xsk3.shape[0])
print(xsk1.shape)

xskv1 = ()
xskv2 = ()
xskv3 = ()
for i in range(x_val.shape[0]):
    xskv1 = np.append(xskv1,scipy.stats.skew(x_val[i,:,:,0].flatten(),axis=0,bias=True))
    xskv2 = np.append(xskv2,scipy.stats.skew(x_val[i,:,:,1].flatten(),axis=0,bias=True))
    xskv3 = np.append(xskv3,scipy.stats.skew(x_val[i,:,:,2].flatten(),axis=0,bias=True))
xskv1=xskv1.reshape(1,xskv1.shape[0])
xskv2=xskv2.reshape(1,xskv2.shape[0])
xskv3=xskv3.reshape(1,xskv3.shape[0])
print(xskv1.shape)

xskt1 = ()
xskt2 = ()
xskt3 = ()
for i in range(test.shape[0]):
    xskt1 = np.append(xskt1,scipy.stats.skew(test[i,:,:,0].flatten(),axis=0,bias=True))
    xskt2 = np.append(xskt2,scipy.stats.skew(test[i,:,:,1].flatten(),axis=0,bias=True))
    xskt3 = np.append(xskt3,scipy.stats.skew(test[i,:,:,2].flatten(),axis=0,bias=True))
xskt1=xskt1.reshape(1,xskt1.shape[0])
xskt2=xskt2.reshape(1,xskt2.shape[0])
xskt3=xskt3.reshape(1,xskt3.shape[0])
print(xskt1.shape)
```

```python
#4. kurtosis

xk1 = ()
xk2 = ()
xk3 = ()
for i in range(x_train.shape[0]):
    xk1 = np.append(xk1,scipy.stats.kurtosis(x_train[i,:,:,0].flatten(),axis=0,bias=True))
    xk2 = np.append(xk2,scipy.stats.kurtosis(x_train[i,:,:,1].flatten(),axis=0,bias=True))
    xk3 = np.append(xk3,scipy.stats.kurtosis(x_train[i,:,:,2].flatten(),axis=0,bias=True))
xk1=xk1.reshape(1,xk1.shape[0])
xk2=xk2.reshape(1,xk2.shape[0])
xk3=xk3.reshape(1,xk3.shape[0])
print(xk3.shape)

xkv1 = ()
xkv2 = ()
xkv3 = ()
for i in range(x_val.shape[0]):
    xkv1 = np.append(xkv1,scipy.stats.kurtosis(x_val[i,:,:,0].flatten(),axis=0,bias=True))
    xkv2 = np.append(xkv2,scipy.stats.kurtosis(x_val[i,:,:,1].flatten(),axis=0,bias=True))
    xkv3 = np.append(xkv3,scipy.stats.kurtosis(x_val[i,:,:,2].flatten(),axis=0,bias=True))
xkv1=xkv1.reshape(1,xkv1.shape[0])
xkv2=xkv2.reshape(1,xkv2.shape[0])
xkv3=xkv3.reshape(1,xkv3.shape[0])
print(xkv1.shape)

xkt1 = ()
xkt2 = ()
xkt3 = ()
for i in range(test.shape[0]):
    xkt1 = np.append(xkt1,scipy.stats.kurtosis(test[i,:,:,0].flatten(),axis=0,bias=True))
    xkt2 = np.append(xkt2,scipy.stats.kurtosis(test[i,:,:,1].flatten(),axis=0,bias=True))
    xkt3 = np.append(xkt3,scipy.stats.kurtosis(test[i,:,:,2].flatten(),axis=0,bias=True))
xkt1=xkt1.reshape(1,xkt1.shape[0])
xkt2=xkt2.reshape(1,xkt2.shape[0])
xkt3=xkt3.reshape(1,xkt3.shape[0])
print(xkt1.shape)
```

```python
# creating feature matrix, f
f_train=np.bmat([[xm11],[xm12],[xm12],[xs1],[xs2],[xs3],[xsk1],[xsk2],[xsk3],[xk1],[xk2],[xk3]])
f_val=np.bmat([[xmv1],[xmv2],[xmv3],[xsv1],[xsv2],[xsv3],[xskv1],[xskv2],[xskv3],[xkv1],[xkv2],[xkv3]])
f_test=np.bmat([[xmt1],[xmt2],[xmt3],[xst1],[xst2],[xst3],[xskt1],[xskt2],[xskt3],[xkt1],[xkt2],[xkt3]])
f_test.shape
f_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
grid={"C":np.logspace(-3,3,7),"penalty":["l1","l2"]},
logistic_regression=LogisticRegression(random_state=100)
log_reg_cv=GridSearchCV(logistic_regression,grid,cv=10)
log_reg_cv.fit(f_train.T,y_train)
```

```python
print("accuracy: ", log_reg_cv.best_score_)
y_pred = log_reg_cv.predict(f_test.T)
y_pred
# from sklearn.metrics import accuracy_score
# accuracy_score(y_val.T, y_pred)
```
